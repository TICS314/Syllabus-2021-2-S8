#<font size="20+"> <font color="blue"> **Histograma (y density plot)**</font>

**¿Qué es un Histograma?**

Un histograma es un tipo de gráfico que nos permite visualizar la distribución de una variable numérica utilizando un conjunto de barras. Cada barra representa un rango determinado de valores, y la altura de una barra en particular representa el número de datos en el DataFrame que pertenecen al rango de valores asociados a esa barra.

**¿Qué es un density plot?**

Es una versión suave del histograma.




#<font color="teal">**Para que sirve los histogramas**</font>
Los histogramas son muy útilies para observar características distribucionales relativas al conjunto de datos. Algunas de las características que podemos observar en los datos son:
1. Simetría
2. Unimodalidad
3. Asimetría hacia la izquierda o derecha
4. Bimodalidad o Multimodalidad
5. Distribución uniforme.


## Distribución simétrica y unimodal:

```python
df=pd.DataFrame(np.random.normal(0,1,1000),columns=["x"])
(
ggplot(df)+
aes(x="x")+
geom_histogram() 
)
```

## Asimetría hacia la izquierda: 

1. La cola de la distribución apunta hacia la izquierda


```python
df=pd.DataFrame(np.random.gamma(2,2,1000)*-1,columns=["x"])
(
ggplot(df)+
aes(x="x")+
geom_histogram() 
)
```

## Asimetría hacia la derecha: 

1. La cola de la distribución apunta hacia la derecha

```python
df=pd.DataFrame(np.random.gamma(3,2,1000),columns=["x"])
(
ggplot(df)+
aes(x="x")+
geom_histogram() 
)
```

## Bimodalidad (Multimodalidad): 

1. Ocurre cuando observamos dos modas en el histograma
2. Los datos multimodales consideran dos o más modas

```python
aux1=np.random.normal(-2,0.5,1000)
aux2=np.random.normal(2,1,1000)


df=pd.DataFrame(np.concatenate((aux1,aux2)),columns=["x"])
(
ggplot(df)+
aes(x="x")+
geom_histogram() 
)
```

## Distribución uniforme

```python
df=pd.DataFrame(np.random.uniform(0,1,10000),columns=["x"])
(
ggplot(df)+
aes(x="x")+
geom_histogram() 
)
```

# <font color="teal"> **Construcción de un histograma**

```python
#Importamos statsmodels.api para importar algunos set de datos
import statsmodels.api as sm

mtcars = sm.datasets.get_rdataset("mtcars", "datasets", cache=True).data
mtcars.head()
```

## Nuestro primer histograma
Construiremos un histograma para la variable ``hp`` relativa a los caballos de fuerza. Para construir un histograma usamos la función  **geom_histogram()**. Esta función recibe argumentos como:
- **color**: controla el color del borde del histograma
- **fill**: controla el color con que pintamos las barras del histograma
- **alpha**: parámetro de transparencia
-**bins**: controla el número de barras.


Además, existen dos tipos de histogramas que podemos construir:
1. **Histograma de frecuencias:** La altura de las barras representa la frecuencia de elementos que pertenecen a dicha barra

```python
(ggplot(mtcars)+
 aes(x="hp")+
 geom_histogram(color="red",fill="red",alpha=0.5)+ 
 theme_linedraw()+
 ggtitle("Histograma Caballos de Fuerza")+
 xlab("Caballos de Fuerza")+
 ylab("Frecuencia")
)
```

2. **Histograma de densidad:** En donde la suma del área de  las barras es igual a 1

Para decirle a ggplot que queremos un gráfico de densidad, le pasamos el siguiente argumento:

**y="..density.."** en **aes()**

```python
(ggplot(mtcars)+
 aes(x="hp",y="..density..")+
 geom_histogram(color="red",fill="red",alpha=0.5)+
 theme_linedraw()+
 ggtitle("Histograma Caballos de Fuerza")+
 xlab("Caballos de Fuerza")+
 ylab("Densidad")
)
```

## **Density plot** 
Además, si queremos añadir una linea correspondiente a la función de densidad podemos usar **geom_density()**. Note que cuando agregamos esta opción el histograma cambia un poco, esto es porque ggplot optimiza la visualización para que el histograma concuerde con la linea de densidad.

```python
(ggplot(mtcars)+
 aes(x="hp",y="..density..")+
 geom_histogram(color="red",fill="red",alpha=0.5)+
 geom_density()+
 scale_x_continuous(limits=[-100,500])+ #Este comando controla el rango del eje x que consideramos en el plot
 theme_linedraw()+
 ggtitle("Histograma Caballos de Fuerza")+
 xlab("Caballos de Fuerza")+
 ylab("Densidad")
)
```

# <font color="teal"> **Dos o más histogramas**

```python
#En este módulo reemplazamos los 0 y 1 por Automático y Manual respectivamente.
indice=mtcars["am"]==0
mtcars["am"][indice]="Automático"

indice=mtcars["am"]==1
mtcars["am"][indice]="Manual"
```

Podemos poner dos o más histogramas (para dos o más grupos) en un mismo gráfico usando los argumentos **fill** y **color** de **aes()**. Estos argumentos van a pintar las barras (el interior y el borde, respectivamente) por grupo. Por defecto, las barras de colores van estar una ecima de la otra, si queremos que se superpongan, debemos usar el argumento **position="indentity"** en **geom_bar()**

```python
(ggplot(mtcars)+
 aes(x="hp",y="..density..",color="am",fill="am")+
 geom_histogram(alpha=0.5,bins=5,position="identity")+
 theme_linedraw()+
 ggtitle("Histograma Caballos de Fuerza")+
 xlab("Caballos de Fuerza")+
 ylab("Densidad")
)
```

Note que el histograma de arriba, no es muy claro, quizas mejores opciónes son
* **Opción 1:** Graficar solamente las densidades


```python
(ggplot(mtcars)+
 aes(x="hp",color="am",fill="am")+
 geom_density(alpha=0.2)+
 theme_linedraw()+
 ggtitle("Histograma Caballos de Fuerza")+
 xlab("Caballos de Fuerza")+
 ylab("Densidad")
)
```

- **Opción 2:**  usar **facet_wrap("am")** que dibuja dos histogramas para cada nivel de la variable **am**.

```python
(ggplot(mtcars)+
 aes(x="hp",color="am",fill="am")+
 geom_histogram(alpha=0.5)+
 theme_linedraw()+
 ggtitle("Histograma Caballos de Fuerza")+
 xlab("Caballos de Fuerza")+
 ylab("Frecuencia")+
 facet_wrap("am")
)
```

También podemos replicar el gráfico de arriba usando **geom_density()** para tener una versión suave de los histogramas.




```python
(ggplot(mtcars)+
 aes(x="hp",color="am",fill="am")+
 geom_density(alpha=0.2)+
 theme_xkcd()+
 ggtitle("Histograma Caballos de Fuerza")+
 xlab("Caballos de Fuerza")+
 ylab("Densidad")+
 facet_wrap("am")+
 scale_fill_manual(name="transmision",values=["violet","teal"])+
 scale_color_manual(name="transmision",values=["violet","teal"])
)
#Note que usando los comandos scale_fill_manual y scale_fill color podemos cambiar los colores
```
